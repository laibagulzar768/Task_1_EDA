4. Data cleaning

```python
data.isnull().sum()
```

***data imputation and removal***

```python
data.drop(columns=['license'], inplace=True)
data.drop(columns=['house_rules'], inplace=True)
```

```python
data.dropna(subset=['NAME'], inplace=True)
data.dropna(subset=['host name'] ,inplace=True)
```

```python



data['host_identity_verified'] = data['host_identity_verified'].fillna(data['host_identity_verified'].mode()[0])
data['instant_bookable'] = data['instant_bookable'].fillna(data['instant_bookable'].mode()[0])
data['neighbourhood group'] = data['neighbourhood group'].fillna(data['neighbourhood group'].mode()[0])
data['neighbourhood'] = data['neighbourhood'].fillna(data['neighbourhood'].mode()[0])
data['country'] = data['country'].fillna(data['country'].mode()[0])
data['country code'] = data['country code'].fillna(data['country code'].mode()[0])
data['lat'] = data['lat'].fillna(data['lat'].median())
data['long'] = data['long'].fillna(data['long'].median())
data['Construction year'] = data['Construction year'].fillna(data['Construction year'].median())
data['minimum nights'] = data['minimum nights'].fillna(data['minimum nights'].median())
data['number of reviews'] = data['number of reviews'].fillna(data['number of reviews'].median())


data.dropna(subset=['price'], inplace=True)
data.dropna(subset=['service fee'], inplace=True)

data['review rate number'] = data['review rate number'].fillna(data['review rate number'].median())
data['calculated host listings count'] = data['calculated host listings count'].fillna(data['calculated host listings count'].median())
data['availability 365'] = data['availability 365'].fillna(data['availability 365'].median())
data['last review'] = data['last review'].fillna("No Review")
data['reviews per month'] = data['reviews per month'].fillna(0)
data['cancellation_policy'] = data['cancellation_policy'].fillna(data['cancellation_policy'].mode()[0])


```

```python
#data check after cleaning
data.isnull().sum()
```

```python
#data after cleaning
data.shape
```

```python
#explore duplicate data
data.duplicated().sum()
data[data.duplicated()]
```

```python
# del duplicated data
data.drop_duplicates(inplace=True)
data.duplicated().sum()
```

```python
#checking datatypes
data.dtypes
data['price'] = pd.to_numeric(data['price'].replace(r'[\$,]', '', regex=True), errors='coerce')
data['service fee'] = pd.to_numeric(data['service fee'].replace(r'[\$,]', '', regex=True), errors='coerce')
```

#### Feature Enginerring

```python
df.groupby(by='neighbourhood group')['price'].mean()
```

```python

df = df[df['price'] < 1200].copy()


df['total_price'] = df['price'] + df['service fee']
df['price_per_night'] = df['total_price'] / df['minimum nights']
df['price_per_available_day'] = df['price'] / df['availability 365']
df['annual_revenue'] = df['price'] * df['availability 365']
```

```python
#aveage price per night 
df.groupby(by='neighbourhood group')['price_per_night'].mean()
```

```python
df['is_verified_host'] = df['host_identity_verified'].map({'verified': 1, 'unverified': 0}) #hot encoding
df['is_multi_listing_host'] = (df['calculated host listings count'] > 1).astype(int) 
df['property_age'] = 2025 - df['Construction year']
df['reviews_per_day'] = df['reviews per month'] / 30
pd.get_dummies(df['room type'], prefix='room')
```

***bivarient analysis***

***Barcharts***

```python
sns.barplot(data=df ,x='neighbourhood group',y='price',hue='room type')
```

```python

avg_price = df.groupby('neighbourhood group')['price_per_night'].mean().sort_values(ascending=False)

sns.barplot(x=avg_price.index, y=avg_price.values)
plt.xticks(rotation=45)
plt.ylabel('Average Price per Night')
plt.title('Average Price per Night by Neighbourhood Group')
plt.show()
```

```python
# Group and sum revenue
revenue_by_room = df.groupby('room type')['annual_revenue'].mean().sort_values(ascending=False)

sns.barplot(x=revenue_by_room.index, y=revenue_by_room.values)
plt.ylabel('Avg Annual Revenue')
plt.title('Annual Revenue by Room Type')
plt.show()
```
